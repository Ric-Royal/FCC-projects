```python
# Create a function grade_marks that given a number will return a grade 
def grade(Total):
    if Total >= 400:
        return 'A'
    elif 350 <= Total <=399:
        return 'B'
    elif 300 <= Total <=349:
        return 'C'
    elif 250 <= Total <=299:
        return 'D'
    else:
        return 'E'

print(grade(200))
```

```python
df['grades'] = df['Total'].apply(grade)

df.head()
```

Plot a line graph commercial imports across the 12 months from 2019 -2021

```python
def plot_df(df, x, y, xlabel='date', ylabel='amount', title='', dpi= 100):
  plt.figure(figsize= (14,12), dpi=dpi)
  plt.plot(x, y, color= 'tab:red')
  plt.gca().set(title= title, xlabel=xlabel, ylabel=ylabel)
  plt.show()

plot_df(ft_summary, x= ft_summary.Date, y=ft_summary['Commercial Imports'], title='Commercial Imports')
```
