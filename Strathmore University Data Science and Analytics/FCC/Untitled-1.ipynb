```python
# Create a function grade_marks that given a number will return a grade 
def grade(Total):
    if Total >= 400:
        return 'A'
    elif 350 <= Total <=399:
        return 'B'
    elif 300 <= Total <=349:
        return 'C'
    elif 250 <= Total <=299:
        return 'D'
    else:
        return 'E'

print(grade(200))
```

```python
df['grades'] = df['Total'].apply(grade)

df.head()
```
